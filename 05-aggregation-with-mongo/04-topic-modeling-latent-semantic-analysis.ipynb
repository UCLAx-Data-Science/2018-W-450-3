```python
import pandas as pd
from pymongo import MongoClient

mongo_client = MongoClient('34.209.155.42', 27016)
database_reference = mongo_client.twitter
```

```python
database_reference.collection_names()
```

```python
collection_reference = database_reference.instructor_test_group
```

```python
all_tweets = {}
just_the_text = {'text': 1}
text = pd.DataFrame(list(collection_reference.find(all_tweets, just_the_text)))
```

```python
text.text = text.text.str.replace('http\S+|www.\S+', '', case=False)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
tfidf.fit(text.text)
word_occurence = tfidf.transform(text.text)
```

```python
word_occurence_m = pd.DataFrame(word_occurence.todense(), columns=tfidf.get_feature_names())
word_occurence_m.head()
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
svd = TruncatedSVD(n_components=50)
```

```python
svd.fit(word_occurence)
```

```python
latent_semantic_analysis = pd.DataFrame(svd.components_, columns=tfidf.get_feature_names()).T
```
